```python
import os,sys
from shutil import copyfile
#copyfile(src, dst)

srcFolderList=[
    "./dataset_new_phishing_360_640",
    "./dataset_new_phishing_360_640_mobile",
    "./dataset_new_phishing_1280_720",
    "dataset_new_phishing_1366_768_trusted",
    "dataset_new_phishing_1920_1080",
]
destFolder="./dataset_new_phishing_mixed"


MAX_IMG_PER_FOLDER=20

for i,srcfolder in enumerate(srcFolderList):
    #srcfolder="./dataset_new_phishing_360_640"
    subfolderlist=os.listdir(srcfolder)
    print(subfolderlist[:10])
    for subfolder in subfolderlist:
        #subfolder = "adobe.com"
        foldername=os.path.join(srcfolder,subfolder)
        print("processing {}".format(foldername))
        if(os.path.isdir(foldername) ==False):#skip non-folder
            continue
        destdir=os.path.join(destFolder,subfolder)
        print("Output dir: {}".format(destdir))
        os.makedirs(destdir, exist_ok=True)
        count=0
        for imgfn in os.listdir(foldername):
            if(imgfn.endswith(".png")):
                img_fullfn=os.path.join(foldername,imgfn)
                #i represents which resolution folder it came from
                destfn=os.path.join(destFolder,subfolder,"{}_{}.png".format(imgfn,i))
                #print(img_fullfn)
                #print(destfn)
                copyfile(img_fullfn,destfn)
                count+=1
                if(count>MAX_IMG_PER_FOLDER):
                    print("got enough stuff in {}".format(destdir))
                    break
                #break
                #end for one resolution, one label
    #break
    #end for one resolution
#end for all
print("done.")
```
